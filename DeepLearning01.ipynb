```python
# Select all rows and columns from index 2 onwards as features (X)
X = data.iloc[:,2:].values

# Select all rows and the column at index 1 ('diagnosis') as the target variable (y)
y = data.iloc[:, 1].values

# Encoding categorical data (diagnosis M/B) into numerical (0/1)
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)

print(f"Shape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split data into training and testing sets
# 10% of the data will be used for testing, 90% for training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Scale the features using StandardScaler
# This transforms the data to have a mean of 0 and a standard deviation of 1
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# adding the input and first hiding layer
classifier = Sequential()
# dense for neuron , unit for outlayer , initilizer = , activatio = activation function  input_din = 30 because we have 30 coloumn
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = 30))

# adding the second hidden layer
classifier.add(Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu'))

# adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

```

```python
classifier.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size=100, epochs=150)
```

```python
#predicting the test set result
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
#
```

```python
#making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm,annot=True)
plt.savefig('confusion_matrix.png')
#True Negatives (Top-Left: 64): These are the cases where the actual diagnosis was Benign (0), and your model correctly predicted it as Benign. Your model correctly identified 64 Benign cases.
# False Positives (Top-Right: 3): These are the cases where the actual diagnosis was Benign (0), but your model incorrectly predicted it as Malignant (1). Your model made 3 false alarms.
# False Negatives (Bottom-Left: 3): These are the cases where the actual diagnosis was Malignant (1), but your model incorrectly predicted it as Benign (0). Your model missed 3 Malignant cases.
# True Positives (Bottom-Right: 44): These are the cases where the actual diagnosis was Malignant (1), and your model correctly predicted it as Malignant. Your model correctly identified 44 Malignant cases.
```
